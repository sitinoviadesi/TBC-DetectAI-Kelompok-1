```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

dataset_path = 'TB_Chest_Radiography_Database'

print("Library berhasil di-import & Path diset!")
```

```python
filepaths = []
labels = []

folds = os.listdir(dataset_path)

for fold in folds:
    foldpath = os.path.join(dataset_path, fold)
    
    # --- BAGIAN PENTING YANG DITAMBAHKAN ---
    # Cek apakah item ini benar-benar Folder? 
    # Kalau bukan (misal file .xlsx), kita 'continue' (lewati)
    if not os.path.isdir(foldpath):
        continue
    # ---------------------------------------
    
    filelist = os.listdir(foldpath)
    
    for file in filelist:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            fpath = os.path.join(foldpath, file)
            filepaths.append(fpath)
            labels.append(fold)

# Buat DataFrame
Fseries = pd.Series(filepaths, name= 'filepaths')
Lseries = pd.Series(labels, name='labels')
df = pd.concat([Fseries, Lseries], axis= 1)

print(f"Total gambar ditemukan: {len(df)}")
print(df.head()) 
print(df['labels'].value_counts())
```

```python
# Split 1: Pisahkan Train (80%) dan Sisa (20%)
train_df, dummy_df = train_test_split(df, train_size=0.8, shuffle=True, random_state=123)

# Split 2: Pisahkan Sisa tadi menjadi Valid (10%) dan Test (10%)
valid_df, test_df = train_test_split(dummy_df, train_size=0.5, shuffle=True, random_state=123)

print(f"Data Latih: {len(train_df)} gambar")
print(f"Data Validasi: {len(valid_df)} gambar")
print(f"Data Uji: {len(test_df)} gambar")
```

```python
# Pengaturan ukuran gambar & batch size
img_size = (224, 224)
batch_size = 32

# 1. Generator untuk Data Latih (Pakai Augmentasi)
tr_gen = ImageDataGenerator(
    rescale=1./255,      # Normalisasi pixel 0-1
    rotation_range=10,   # Putar gambar max 10 derajat
    width_shift_range=0.1, # Geser horizontal
    height_shift_range=0.1, # Geser vertikal
    zoom_range=0.1,      # Zoom in/out
    horizontal_flip=True # Balik kiri-kanan
)

# 2. Generator untuk Validasi & Test (JANGAN pakai Augmentasi, hanya rescale)
ts_gen = ImageDataGenerator(rescale=1./255)

# Load data ke dalam generator
train_gen = tr_gen.flow_from_dataframe(train_df, x_col='filepaths', y_col='labels', target_size=img_size, class_mode='binary', color_mode='rgb', shuffle=True, batch_size=batch_size)
valid_gen = ts_gen.flow_from_dataframe(valid_df, x_col='filepaths', y_col='labels', target_size=img_size, class_mode='binary', color_mode='rgb', shuffle=True, batch_size=batch_size)
test_gen = ts_gen.flow_from_dataframe(test_df, x_col='filepaths', y_col='labels', target_size=img_size, class_mode='binary', color_mode='rgb', shuffle=False, batch_size=batch_size)
```

```python
# Ambil 1 batch gambar dari training generator
g_dict = train_gen.class_indices      # Cek label (mana 0, mana 1)
classes = list(g_dict.keys())
images, labels = next(train_gen)      # Ambil sampel

plt.figure(figsize=(15, 15))

for i in range(16): # Tampilkan 16 gambar
    plt.subplot(4, 4, i + 1)
    image = images[i]
    plt.imshow(image)
    
    # Tampilkan label (Normal/Tuberculosis)
    index = int(labels[i])
    class_name = classes[index]
    plt.title(class_name, color='blue', fontsize=12)
    plt.axis('off')

plt.show()
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# 1. Download Base Model (MobileNetV2) tanpa bagian kepalanya (include_top=False)
# 'imagenet' artinya kita pakai bobot yang sudah terlatih mengenali jutaan gambar dunia nyata
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 2. Bekukan Base Model (agar ilmu lamanya tidak hilang saat dilatih)
base_model.trainable = False 

# 3. Tambahkan "Kepala" Baru (Custom Layers) untuk kasus TBC kita
x = base_model.output
x = GlobalAveragePooling2D()(x) # Meratakan fitur
x = Dense(128, activation='relu')(x) # Layer berpikir (128 neuron)
x = Dropout(0.2)(x) # Mematikan 20% neuron secara acak biar tidak menghafal (Overfitting)

# Output Layer: Cuma butuh 1 neuron karena klasifikasinya Binary (TBC vs Normal)
# Activation 'sigmoid' akan menghasilkan angka 0 sampai 1 (Peluang)
predictions = Dense(1, activation='sigmoid')(x) 

# 4. Gabungkan menjadi satu model utuh
model = Model(inputs=base_model.input, outputs=predictions)

# 5. Kompilasi Model
model.compile(optimizer=Adam(learning_rate=0.0001), 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

# Tampilkan ringkasan model
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# 1. Siapkan Callbacks
# Stop kalau akurasi validasi gak naik-naik selama 3 epoch (patience=3)
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Simpan model terbaik ke file 'tbc_model.h5'
checkpoint = ModelCheckpoint('tbc_model.h5', monitor='val_accuracy', save_best_only=True)

# 2. Mulai Training
epochs = 15  # Kita coba 15 putaran dulu

history = model.fit(
    train_gen,
    validation_data=valid_gen,
    epochs=epochs,
    callbacks=[early_stop, checkpoint],
    verbose=1
)
```

```python
# Ambil data dari hasil training tadi
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(len(acc)) # Menyesuaikan jumlah epoch yang jalan (misal 5)

plt.figure(figsize=(15, 5))

# --- Plot 1: Grafik Akurasi ---
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Grafik Akurasi (Semakin Tinggi Semakin Baik)')
plt.grid(True)

# --- Plot 2: Grafik Loss (Error) ---
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Grafik Loss/Error (Semakin Rendah Semakin Baik)')
plt.grid(True)

plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# 1. Ambil Prediksi dari Data Test
# Penting: shuffle=False di test_gen (Cell 4) sangat krusial di sini
print("Sedang melakukan prediksi pada data test...")
predictions = model.predict(test_gen)
y_pred = (predictions > 0.5).astype(int) # Ubah probabilitas jadi 0 atau 1
y_true = test_gen.classes # Label asli (kunci jawaban)

# 2. Buat Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

# 3. Visualisasi Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'TBC'], yticklabels=['Normal', 'TBC'])
plt.xlabel('Prediksi Model')
plt.ylabel('Kenyataan (Label Asli)')
plt.title('Confusion Matrix: Deteksi TBC')
plt.show()

# 4. Tampilkan Laporan Lengkap (Precision, Recall, F1-Score)
print("\nLaporan Klasifikasi:")
print(classification_report(y_true, y_pred, target_names=['Normal', 'Tuberculosis']))
```

```python
from tensorflow.keras.preprocessing import image

# --- GANTI PATH INI DENGAN SALAH SATU GAMBAR UNTUK DITES ---
# Pilih gambar yang belum pernah dilihat model (misal ambil dari Google atau folder Test)
path_gambar_tes = r'C:/laragon/www/Tuberculosis/TB_Chest_Radiography_Database//Normal/Normal-1.png'
# -----------------------------------------------------------

def prediksi_gambar(path_file):
    try:
        # 1. Load dan Preprocess gambar (sama persis dengan saat training)
        img = image.load_img(path_file, target_size=(224, 224))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) # Ubah jadi batch (1, 224, 224, 3)
        img_array = img_array / 255.0 # Normalisasi
        
        # 2. Prediksi
        score = model.predict(img_array)[0][0]
        
        # 3. Tampilkan Hasil
        plt.imshow(img)
        plt.axis('off')
        
        if score > 0.5:
            label = "TERINDIKASI TUBERKULOSIS (TBC)"
            prob = score * 100
            color = 'red'
        else:
            label = "PARU-PARU NORMAL"
            prob = (1 - score) * 100
            color = 'green'
            
        plt.title(f"{label}\nKeyakinan Model: {prob:.2f}%", color=color, fontweight='bold')
        plt.show()
        
    except Exception as e:
        print(f"Error: {e}. Pastikan path file benar!")

# Jalankan fungsi
prediksi_gambar(path_gambar_tes)
```
